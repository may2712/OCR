```python
import pytesseract
from PIL import Image
import cv2
import numpy as np
import language_tool_python
from google.colab import files

print("Téléverse ton image :")
uploaded = files.upload()
filename = next(iter(uploaded))

file_bytes = uploaded[filename]
np_arr = np.frombuffer(file_bytes, np.uint8)
image = cv2.imdecode(np_arr, cv2.IMREAD_COLOR)
if image is None:
    raise ValueError("Erreur lors du chargement de l'image")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

pil_img = Image.fromarray(gray)

texte = pytesseract.image_to_string(pil_img, lang='fra')
print("=== Texte brut extrait ===\n", texte)

tool = language_tool_python.LanguageTool('fr')
matches = tool.check(texte)
texte_corrige = language_tool_python.utils.correct(texte, matches)

print("\n=== Texte corrigé automatiquement ===\n", texte_corrige)
```
